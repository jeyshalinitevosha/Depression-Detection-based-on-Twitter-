```python
#import modules
import twint
import nest_asyncio
nest_asyncio.apply()
import pandas as pd
import re
```

```python
#configure TWINT
t = twint.Config()

t.Search = "depressed", "lonely"
t.Lang = "en"
t.Limit = 5000
t.Year = 2020
t.Store_csv = True
t.Output = "data/tweets_depressed.csv"

twint.run.Search(t)
```

```python
df = pd.read_csv("data/tweets_depressed.csv")
len(df)
```

```python
df.hashtags.value_counts()
```

```python
#remove duplicates
df = df.drop_duplicates(subset = ["id"])
df.shape
```

```python
#remove unwanted/postive tweet entries 
selection_to_remove = ["travel", "wanderlust", "adventure", "explore", "yoga",
                      "yogi", "backpacking", "backpacker", "poetry", "izabellkey",
                      "quotestoliveby", "wisdom", "freshmindclinic", "depressionhelp",
                      "connection", "twitch", "happyholidays2019", "endofdecadelgbtq"]
```

```python
mask1 = df.hashtags.apply(lambda x:any(item for item in selection_to_remove 
                                       if item in x))
df[mask1].tweet.tail()
```

```python
#review results
df[mask1==False].tweet.head(10)
```

```python
#apply mask1
df = df[mask1 == False]
len(df)
```

```python
#remove entries with more than three hashtags
mask2 = df.tweet.apply(lambda x: x.count("#") < 4)
```

```python
#apply mask2
df = df[mask2]
len(df)
```

```python
#remove tweets with mentions as they may be retweets
mask3 = df.mentions.apply(lambda x: len(x) < 5 )
```

```python
#apply mask3
df = df[mask3]
len(df)
```

```python
df.hashtags.value_counts()
```

```python
#remove unwanted/postive tweet entries 
selection_to_remove = ["justkidding", "famcommunity", "praying", "jesuswasbornforyou",
                      "youareworthy"]
```

```python
mask4 = df.hashtags.apply(lambda x:any(item for item in selection_to_remove 
                                       if item in x))
df[mask4].tweet.tail()
```

```python
#review results
df[mask4==False].tweet.head(10)
```

```python
#apply mask4
df = df[mask4 == False]
len(df)
```

```python
#remove entries with URLs
mask5 = df.urls.apply(lambda x: len(x) < 5)
```

```python
#review dataset
df[mask5 == False].tweet.head(10), df[mask5 == False].tweet.tail(10)
```

```python
#apply mask5
df = df[mask5]
len(df)
```

```python
#create a column containing tweets without hashtags
df["mod_text"] = df["tweet"].apply(lambda x: re.sub(r'#\w+', '', x))
```

```python
df.mod_text.head(15), df.mod_text.tail(15)
```

```python
df.columns
```

```python
col_list = ["id", "conversation_id", "date", "username", "mod_text", "hashtags", "tweet"]
```

```python
df_final = df[col_list]
df_final = df_final.rename(columns={"mod_text": "tweet_processed", "tweet": "tweet_original"})
```

```python
df_final["target"] = 1
```
